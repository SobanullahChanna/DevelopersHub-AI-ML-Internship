```python
# Task 1: Iris Data Exploration
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
# Seaborn has this dataset built-in, so it's very easy!
df = sns.load_dataset('iris')

print("✅ Data Loaded Successfully!")
```

```python
# Print the shape (Rows, Columns)
print(f"Shape of dataset: {df.shape}")

# Print column names
print(f"Column names: {df.columns.tolist()}")

# Show the first 5 rows
df.head()
```

```python
# Show data types and missing values
print("--- Data Info ---")
df.info()

# Show math summary (mean, min, max, etc.)
print("\n--- Summary Statistics ---")
df.describe()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', style='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
df.hist(edgecolor='black', linewidth=1.2, figsize=(10, 8))
plt.suptitle("Distribution of Iris Features")
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(data=df, x='species', y='petal_length')
plt.title('Petal Length Distribution by Species')
plt.show()
```
