```python
import pandas as pd
import os

# Create the data folder if it doesn't exist
if not os.path.exists('data'):
    os.makedirs('data')

# The "address" where the heart data lives online
url = "https://raw.githubusercontent.com/amankharwal/Website-data/master/heart.csv"

# Tell Python to go get it and save it as 'heart.csv'
df = pd.read_csv(url)
df.to_csv('data/heart.csv', index=False)

print("✅ Success! Your heart.csv file is now sitting in the 'data' folder.")
df.head() # Show the first few rows to make sure it's correct
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the file we just saved
df = pd.read_csv('data/heart.csv')

# Create a simple count plot
sns.countplot(x='target', data=df, palette='magma')
plt.title('Heart Disease: Healthy (0) vs. Sick (1)')
plt.show()
```
